```python
import numpy as np
```

```python
from multiprocessing import Pool, set_start_method, get_context
```

```python
def f(x):
    return x*x

def main():
    with Pool(5) as p:
        print(p.map(f, [1, 2, 3]))
```

```python
if __name__ == '__main__':
    ctx = get_context('spawn')  # Explicitly set the spawn method
    with Pool(3) as p:
        print(p.map(f, [1, 2, 3]))
```
